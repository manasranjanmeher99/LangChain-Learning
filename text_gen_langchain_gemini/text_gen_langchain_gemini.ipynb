```python
import os
from langchain.chat_models import init_chat_model

os.environ["GOOGLE_API_KEY"]= ""

model= init_chat_model("google_genai:gemini-3.5-flash")
```

```python
response= model.invoke("about software jobs in 2026 to 2030")
```

```python
print(response)
```

```python
print(response.text)
```
